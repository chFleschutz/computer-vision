# Load Images

```python
import os

image_directory = 'pedestrian-dataset/PNGImages'
image_files = [f for f in os.listdir(image_directory) if f.endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

print(image_files.__len__())
```

```python
import cv2

image_path = os.path.join(image_directory, image_files[0])
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)
```

```python
import matplotlib.pyplot as plt

plt.imshow(image)
```

```python
grey_imge = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
boxes, weights = hog.detectMultiScale(grey_imge, winStride=(2, 2))

debug_image = image.copy()
for (x, y, w, h) in boxes:
    cv2.rectangle(debug_image, (x, y), (x + w, y + h), (0, 255, 0), 2)

plt.imshow(debug_image)
```
